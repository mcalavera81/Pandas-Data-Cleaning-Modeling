```python
ebola= pd.read_csv('./country_timeseries.csv', parse_dates=['Date'], index_col='Date')
ebola.head()
```

```python
new_idx  = pd.date_range(ebola.index.min(), ebola.index.max())
new_idx
```

```python
ebola.index
```

```python
new_idx = reversed(new_idx)
```

```python
ebola = ebola.reindex(new_idx)
ebola.head()
```

```python
ebola['Cases_Liberia']
```

```python
last_valid = ebola.apply(pd.Series.last_valid_index)
last_valid
```

```python
earliest_data = ebola.index.min()
earliest_data
```

```python
shift_values = last_valid -earliest_data
shift_values
```

```python
ebola_dict = {}

for idx , col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
```

```python
ebola_shift = pd.DataFrame(ebola_dict, columns=ebola.columns)
ebola_shift.tail()
```

```python
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Day'],axis=1)
ebola_shift
```
